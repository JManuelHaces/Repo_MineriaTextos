## Buscando palabras similares

```python
text = nltk.Text(word.lower() for word in nltk.corpus.brown.words())
```

```python
# Busca palabras similares a la dada
text.similar('woman')
```

- Lo anterior es posible buscando tripletas, en donde la primera y la última palabra sea igual a las que están con las buscadas, por lo que la de en medio es similar a ella.



- wn = Cualquier palabra 
- w2 = Palabra buscada
- w2' = Palabra similar
    - $Tripleta1 = w1 w2 w3$
    - $Tripleta2 = w1 w2' w3$
    - $si w1 == w2 and w3 == w3 -> entonces w2 == w2'$

## ¿Cuáles son los tipos de palabras (etiquetas) que más aparecen en un texto?

```python
from nltk.corpus import brown

# Haciendo un conteo de etiquetas en un corpus
brown_new_tagged = brown.tagged_words(categories='news', tagset='universal')
tag_fd = nltk.FreqDist(tag for (word, tag) in brown_new_tagged)
tag_fd.most_common()
```

```python
# Grpafica acumulada
tag_fd.plot(cumulative=True)
```

## ¿Cuál es la etiqueta que más aparece antes de un sustativo ('noun')?

```python
# Obteniendo bigramas del corpus etiquetado
word_tag_pairs = nltk.bigrams(brown_new_tagged)

# Haciendo una lista de tuplas con las palabras que preceden a un sustativo (NOUN)
noun_preceders = [a[1] for (a, b) in word_tag_pairs if b[1] == 'NOUN']

# Haciendo una distribución de frecuencia
fdist = nltk.FreqDist(noun_preceders)

# Haciendo una lista del más frecuente al menos frecuente
[tag for (tag, _) in fdist.most_common()]
```

## Etiquetando un Artículo de nuestro Corpus
- Se puede etiquetar cualquier documento que esté en ingles.
- Para este caso se usará uno de los archivos txt obtenidos de la práctica 1.

```python
with open('00.txt', 'r') as f:
    lines = f.readlines()
```

```python
text = ''
annotated_text = []

for line in lines:
    text = nltk.word_tokenize(line)
    annotated_text.append(nltk.pos_tag(text, tagset='universal'))
```

```python
# Extraer en una lista todos los sustantivos 
sustantivos = []

for line in annotated_text:
    for word in line:
        if word[1] == 'NOUN':
            sustantivos.append(word[0])
```
